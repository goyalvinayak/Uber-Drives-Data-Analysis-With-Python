```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime, time

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df = pd.read_csv('Dataset/My Uber Drives.csv')
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace("*","")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['PURPOSE'].fillna(method = 'ffill', inplace = True)
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0, subset=['END_DATE','CATEGORY','START','STOP'], how='all', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['START_DATE'] = pd.to_datetime(df['START_DATE'], errors='coerce')
df['END_DATE'] = pd.to_datetime(df['END_DATE'], errors='coerce')
```

```python
df.info()
```

```python
df['CATEGORY'].value_counts()
```

```python
sns.countplot(x=df['CATEGORY']);
```

```python
start_labels = df.START.value_counts().nlargest(10)
sns.barplot(x=start_labels.index, y=start_labels);
plt.ylabel('Value Counts');
plt.xticks(rotation=90);
```

```python
df.PURPOSE.value_counts().plot(kind = 'bar')
```

```python
def round(x):
    if x['START'] == x['STOP']:
        return 'Yes'
    else:
        return 'No'
```

```python
df['ROUND_TRIP'] = df.apply(round, axis = 1)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df[ 'MILES' ].plot.hist()

plt.xlabel("Miles Range")
plt.ylabel("Number of Trips")

# setting y_label as price
plt.title("Number of Trips by Miles")
plt.show()
```

```python
df['ROUND_TRIP'].value_counts().plot(kind='bar')
```

```python
df['MONTH'] = pd.DatetimeIndex(df['START_DATE']).month
```

```python
dict = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'July', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
```

```python
df['MONTH'] = df['MONTH'].map(dict)
```

```python
df.head()
```

```python
df['MONTH'].value_counts().plot(kind='pie')
```

```python
month_count = df.MONTH.value_counts()
sns.barplot(x=month_count.index, y=month_count);
plt.xlabel('Months')
plt.ylabel('Value Counts');
```

```python
df['DAY'] = df.START_DATE.dt.weekday
day_label = {
    0: 'Mon', 1: 'Tues', 2: 'Wed', 3: 'Thus', 4: 'Fri', 5: 'Sat', 6: 'Sun'
}
df['DAY'] = df['DAY'].map(day_label)
```

```python
day_label = df.DAY.value_counts()
sns.barplot(x=day_label.index, y=day_label);
plt.xlabel('DAY')
plt.ylabel('COUNT');
```

```python
sns.countplot(hue='CATEGORY', x='DAY', data=df);
```
